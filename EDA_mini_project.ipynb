```python
#Value,Wage,Joined,Contract Valid Until,Height,Weight,Release Clause
```

```python
fifa["Value"]= fifa["Value"].str.replace("€","")
fifa["Value"]= fifa["Value"].str.replace("M","")
fifa["Value"]= fifa["Value"].str.replace("K","")
fifa['Value']=pd.to_numeric(fifa['Value'])
```

```python
fifa["Wage"]= fifa["Wage"].str.replace("€","")
fifa["Wage"]= fifa["Wage"].str.replace("K","")
fifa['Wage']=pd.to_numeric(fifa['Wage'])
```

```python
fifa["Release Clause"]= fifa["Release Clause"].str.replace("€","")
fifa["Release Clause"]= fifa["Release Clause"].str.replace("K","")
fifa["Release Clause"]= fifa["Release Clause"].str.replace("M","")
fifa['Release Clause']=pd.to_numeric(fifa['Release Clause'])
```

```python
fifa['Joined']=pd.to_datetime(fifa['Joined'])
fifa['Joined']=fifa['Joined'].dt.year
fifa['Joined']
```

```python
fifa['Contract Valid Until']=pd.to_datetime(fifa['Contract Valid Until'],yearfirst=True)
fifa['Contract Valid Until']
```

```python
fifa["Height"]= fifa["Height"].str.replace("\'",".")
fifa['Height']=pd.to_numeric(fifa['Height'])
```

```python
fifa["Weight"]= fifa["Weight"].str.replace("lbs",".")
fifa['Weight']=pd.to_numeric(fifa['Weight'])
```

```python
fifa.isnull().sum()
```
